```python
import pandas as pd
import statistics
```

```python
waste = pd.read_csv('/home/dside/Documents/GitHub/MunicipalData/Monthly/fruitless_expenditure.csv')

IncExp = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Monthly/income_expenditure.csv")
OpExp = IncExp[IncExp['label'] == 'Total Operating Expenditure']
OpExp = OpExp[OpExp['date'] != 2016] # dont have 2016 data for waste
#OpExp = OpExp.rename(columns = {'Unnamed: 0': 'date'})
```

```python
OpExp = OpExp.sort_values('date')
OpExp
```

```python
waste
```

```python
fruitless = waste[waste['label'] == 'Fruitless and Wasteful Expenditure']
irregular = waste[waste['label'] == 'Irregular Expenditure']
unauthorised = waste[waste['label'] == 'Unauthorised Expenditure']
```

```python
unauthorised
```

Notice that some of the wasted expenditure was NaN. These values were replaced by zeros

```python
Efficiency = pd.DataFrame()

Efficiency = fruitless[['date', 'mun_code', 'total_amount']]
Efficiency = Efficiency.rename(columns = {'total_amount': 'fruitless'})
Efficiency = Efficiency.merge(irregular, on = ['date', 'mun_code'])
Efficiency = Efficiency.rename(columns = {'total_amount': 'irregular'})
Efficiency= Efficiency[['date', 'mun_code', 'fruitless', 'irregular']]

Efficiency = Efficiency.merge(unauthorised, on = ['date', 'mun_code'])
Efficiency = Efficiency.rename(columns = {'total_amount': 'unauthorised'})

Efficiency = Efficiency[['date', 'mun_code', 'fruitless', 'irregular', 'unauthorised']]
Efficiency['total_waste'] = Efficiency['fruitless'] + Efficiency['irregular'] + Efficiency['unauthorised']

Efficiency = Efficiency.merge(OpExp, on = ['date', 'mun_code'])
Efficiency = Efficiency.rename(columns = {"total_amount":"OpExp"})
Efficiency = Efficiency[['date', 'mun_code', 'fruitless', 'irregular', 'unauthorised', 'total_waste', 'OpExp']]

Efficiency.fillna(0)


Efficiency
```

```python
# THE BIGGER THIS MEASURE THE MORE EFFICIENT THE SPENDING

Efficiency['Efficient_ratio'] = Efficiency['total_waste'] / Efficiency['OpExp']
Efficiency
```

```python
Eff_mean = statistics.mean(Efficiency['Efficient_ratio'])

Eff_max = max(Efficiency['Efficient_ratio'])
Eff_min = min(Efficiency['Efficient_ratio'])
Eff_sd = statistics.stdev(Efficiency['Efficient_ratio'])

#normalising the efficiency measure
Efficiency['Efficient_ratio'] = (Efficiency['Efficient_ratio'] - Eff_mean) / Eff_sd
#Efficiency['Efficient_ratio'] = Efficiency['Efficient_ratio'] - 0.5 

Efficiency['Efficient_measure'] = 1 - Efficiency['Efficient_ratio'] 

Efficiency
```

```python
# Wellbeing mesure

wellbeing = pd.read_csv('/home/dside/Documents/GitHub/Youth data/YOUTH_WELFARE.csv')
wellfare_mean = statistics.mean(wellbeing['wellfare_measure'])
wellfare_std = statistics.stdev(wellbeing['wellfare_measure'])

wellbeing
```

```python
wellbeing['wellfare_measure'] = (wellbeing['wellfare_measure'] - wellfare_mean) / wellfare_std 
wellbeing
```

```python
Eff_Well = Efficiency.merge(wellbeing, how = 'left', on = ['mun_code'])
Eff_Well
```

```python
Eff_Well_temp = Eff_Well.dropna()
Eff_Well_temp.shape # 933 X 15
```

```python
# Calculating the profiles

profile=[]

for i in range(0,len(Eff_Well_temp)):
    if Eff_Well_temp['Efficient_ratio'].iloc[i] >= 0 and Eff_Well_temp['wellfare_measure'].iloc[i] >= 0:
        profile.append(1)
    if Eff_Well_temp['Efficient_ratio'].iloc[i] >= 0 and Eff_Well_temp['wellfare_measure'].iloc[i] < 0:
         profile.append(2)
    if Eff_Well_temp['Efficient_ratio'].iloc[i] < 0 and Eff_Well_temp['wellfare_measure'].iloc[i] < 0:
         profile.append(3)
    if Eff_Well_temp['Efficient_ratio'].iloc[i] < 0 and Eff_Well_temp['wellfare_measure'].iloc[i] >= 0:
         profile.append(4)
    
            
```

```python
profile_series = pd.Series(profile)
Eff_Well_temp['profile'] = profile_series.values

Eff_Well_temp = Eff_Well_temp[['date', 'mun_code', 'Efficient_measure', 'wellfare_measure', 'province', 'profile']]
Eff_Well_temp

```
