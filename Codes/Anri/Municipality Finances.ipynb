## current ratio 

\frac{current assets}{current liabilities}

```python
# Current Ratio
balSheet = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/balancesheet_total.csv")
balSheet
```

```python
currAssets = balSheet[balSheet["label"] == "Total Current Assets"]
currAssets = currAssets[['Unnamed: 0', 'mun_name','mun_code','label', 'total_amount']]
currAssets.sort_values('mun_code')
currAssets.head()
```

```python
currLiabilities = balSheet[balSheet["label"] == "Total Current Liabilities"]
currLiabilities = currLiabilities[['Unnamed: 0','mun_name','mun_code','label', 'total_amount']]
currLiabilities.sort_values('mun_code')
#currLiabilities.head()

#currLiabilities[currLiabilities["mun_code"] == "BUF"]
```

```python
result = currAssets.merge(currLiabilities, on = 'mun_code', how = 'left')
result

result = result[['Unnamed: 0_x', 'mun_name_x', 'mun_code' ,'total_amount_x','total_amount_y']]
result = result.rename(columns = {'mun_name_x': 'mun_name' , 'total_amount_x': 'curr_assets', 'total_amount_y':'curr_liab'  })

#calculating curr ratio
result['currRatio'] = result['curr_assets']/result['curr_liab']
result
```

## cash / cost coverage 

This ration is used to indicate the municipalities ability to meet its operating commitments from cach and short-term investment without collecting any additional revenue.

\frac{cash - unspent conditional grants - overdraft + short-term investment}{Operational expenditure}

note:conditional grants, short-term investment still needs to be added

```python
cash = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Cash_flow_total.csv")
cash.head()
```

```python
#Cash receipts

cash_receipts = cash[cash["label"] == "Total Cash Receipts by Source"]
cash_receipts = cash_receipts[['Unnamed: 0', 'mun_code', 'total_amount']]
cash_receipts = cash_receipts.rename(columns = {'total_amount': 'cash_rec'})
cash_receipts.head()
```

```python
#cash payments

cash_payments = cash[cash["label"] == "Total Cash Payments by Type"]
cash_payments = cash_payments [['Unnamed: 0','mun_code', 'total_amount']]
cash_payments = cash_payments.rename(columns = {'total_amount': 'cash_pay'})
cash_payments.head()
```

```python
#cash balance

result = cash_receipts.merge(cash_payments, on = ['mun_code', 'Unnamed: 0'], how = 'left')
result['cash_bal'] = result['cash_rec'] - result['cash_pay']
result.sort_values('mun_code')

result # 1052 x 5
```

```python
# overdraft

balsheetRAW = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Balancesheet.csv")
balsheetRAW

overdraft = balsheetRAW[balsheetRAW["label"] == "Bank Overdraft"]
overdraft.sort_values('mun_code')
overdraft #831 x 6
```

```python
balsheetRAW.shape
balsheetRAW.iloc[34070,0]
```

```python
# calculating the measure

# note waiting on updated balance sheet (missing 2016)
cash_costData = overdraft.merge(result, on = ['mun_code', 'Unnamed: 0'], how = 'left')
cash_costData =  cash_costData[['Unnamed: 0', 'mun_name', 'mun_code', 'total_amount', 'cash_bal']]
cash_costData = cash_costData.rename(columns = {'total_amount': 'overdraft'})


cash_costData = cash_costData.merge(OpExp, on = ['mun_code', 'Unnamed: 0', 'mun_name'], how = 'left')
del cash_costData['label']
cash_costData = cash_costData.rename(columns = {'total_amount': 'opExp'})
cash_costData

cash_costCoverage = (cash_costData['cash_bal'] - cash_costData['overdraft'] ) / cash_costData['opExp']
cash_costCoverage
```

## Fruitless and irregular expenditure 

This ratio has been formulated to compare the fruitless expenditure to the total expenditure.


\frac{fruitless Expenditure + irregular Expenditure}{Operational expenditure}


```python
wastedExp = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Wasteful_expenditure.csv")
wastedExp.sort_values("mun_code")
```

```python
wastedExp = wastedExp.groupby(by = ['mun_code', 'Unnamed: 0']).sum()
#wastedExp #833 x 1
```

```python
wastedProp = pd.DataFrame()

wastedProp['total'] = OpExpGrouped['total_amount']

wastedProp = wastedProp.merge(wastedExp, how = 'left') 

wastedProp


#wastedExp = wastedExp.merge(OpExp, on = ['mun_code', 'Unnamed: 0', 'mun_name'], how = 'left')
#wastedExp
```
