## cash / cost coverage 

This ration is used to indicate the municipalities ability to meet its operating commitments from cash and short-term investment without collecting any additional revenue.

\frac{cash - unspent conditional grants - overdraft + short-term investment}{Operational expenditure}

note:conditional grants, short-term investment still needs to be added

```python
cash = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Cash_flow_total.csv")
cash.head()
```

```python
#Cash receipts

cash_receipts = cash[cash["label"] == "Total Cash Receipts by Source"]
cash_receipts = cash_receipts[['Unnamed: 0', 'mun_code', 'total_amount']]
cash_receipts = cash_receipts.rename(columns = {'total_amount': 'cash_rec'})
cash_receipts.head()
```

```python
#cash payments

cash_payments = cash[cash["label"] == "Total Cash Payments by Type"]
cash_payments = cash_payments [['Unnamed: 0','mun_code', 'total_amount']]
cash_payments = cash_payments.rename(columns = {'total_amount': 'cash_pay'})
cash_payments.head()
```

```python
#cash balance

result = cash_receipts.merge(cash_payments, on = ['mun_code', 'Unnamed: 0'], how = 'left')
result['cash_bal'] = result['cash_rec'] - result['cash_pay']
result.sort_values('mun_code')

result # 1052 x 5
```

```python
# overdraft

balsheetRAW = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Balancesheet.csv")
balsheetRAW

overdraft = balsheetRAW[balsheetRAW["label"] == "Bank Overdraft"]
overdraft.sort_values('mun_code')
overdraft #831 x 6
```

```python
# calculating the measure

# note waiting on updated balance sheet (missing 2016)
cash_costData = overdraft.merge(result, on = ['mun_code', 'Unnamed: 0'], how = 'left')
cash_costData =  cash_costData[['Unnamed: 0', 'mun_name', 'mun_code', 'total_amount', 'cash_bal']]
cash_costData = cash_costData.rename(columns = {'total_amount': 'overdraft'})


cash_costData = cash_costData.merge(OpExp, on = ['mun_code', 'Unnamed: 0', 'mun_name'], how = 'left')
del cash_costData['label']
cash_costData = cash_costData.rename(columns = {'total_amount': 'opExp'})
cash_costData

cash_costCoverage = (cash_costData['cash_bal'] - cash_costData['overdraft'] ) / cash_costData['opExp']
cash_costCoverage
```

## Net Operating Surplus Margin 

The Ratio assesses the extent to which the Municipality generates Operating Surpluses. This is an efficiency measure.


```python
IncExp = pd.read_csv("/Users/anrismith/Downloads/DSIDE-master/Municipal data/Income_Expenditure_total.csv")

OpRev = IncExp[IncExp['label'] == 'Total Operating Revenue']
OpExp = IncExp[IncExp['label'] == 'Total Operating Expenditure']


OpRev = OpRev[['Unnamed: 0', 'mun_code', 'total_amount']]
OpRev = OpRev.rename(columns = {'total_amount': 'opRev'})
OpExp = OpExp[['Unnamed: 0', 'mun_code', 'total_amount']]
OpExp = OpExp.rename(columns = {'total_amount': 'opExp'})

OpFinal = OpRev.merge(OpExp, on = ['mun_code', 'Unnamed: 0'])
```

```python
OpFinal['OppSurplusMargin'] = (OpFinal['opRev'] - OpFinal['opExp']) / OpFinal['opRev']
OpFinal
```

## Fruitless and irregular expenditure - Efficient spending measure

This ratio has been formulated to compare the fruitless expenditure to the total expenditure.


\frac{fruitless Expenditure + irregular Expenditure}{Operational expenditure}


```python
wastedExp = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Wasteful_expenditure.csv")
#wastedExp #wasted Exp only till 2015

OpExp = OpExp[OpExp['Unnamed: 0'] != 2016]
```

```python
fruitless = wastedExp[wastedExp['label'] == 'fruitless']
irregular = wastedExp[wastedExp['label'] == 'irregular']
unauthorised = wastedExp[wastedExp['label'] == 'unauthorised']
```

```python
waste = pd.DataFrame()
#waste['date'] = wastedExp['Unnamed: 0']
#waste['mun_code'] = wastedExp['mun_code']

waste = fruitless[['Unnamed: 0', 'mun_code', 'total_amount']]
waste = waste.rename(columns = {'total_amount': 'fruitless'})
waste = waste.merge(irregular, on = ['Unnamed: 0', 'mun_code'])
waste = waste.rename(columns = {'total_amount': 'irregular'})

waste = waste[['Unnamed: 0', 'mun_code', 'fruitless', 'irregular' ]]

waste = waste.merge(unauthorised, on = ['Unnamed: 0', 'mun_code'])
waste = waste.rename(columns = {'total_amount': 'unauthorised'})

waste = waste[['Unnamed: 0', 'mun_code', 'fruitless', 'irregular', 'unauthorised']]

waste['total_waste'] = waste['fruitless'] + waste['irregular'] + waste['unauthorised']

waste = waste.merge(OpExp, on = ['Unnamed: 0', 'mun_code'])
waste = waste.rename(columns = {"total_amount":"OpExp"})

# THE BIGGER THIS MEASURE THE MORE EFFICIENT THE SPENDING
waste['Efficient_spending'] = 1 - waste['total_waste'] / waste['OpExp']
waste
```

## repairs and maintenance as a percentage of  PPE

This measure assists in determining the level of repairs and maintenance to ensure adequate
maintenance to prevent breakdowns and interruptions to service delivery:

Total Repairs and Maintenance Expenditure / Property, Plant and Equipment and
Investment Property (Carrying Value) x 100

```python
#import pandas as pd
repairs = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Repair_Maintance_total.csv")
#repairs.head()
```

```python
# Repairs and maintenance
repairs_PPE = repairs[['Unnamed: 0', 'mun_code', 'mun_name', 'Province', 'Latitude', 'Longitude', 'total_amount']]
repairs_PPE = repairs_PPE.rename(columns = {'total_amount': 'repairs_maintenance'})
repairs_PPE.head()
```

```python
#PPE
balsheet = pd.read_csv("/home/dside/Documents/GitHub/MunicipalData/Balancesheet.csv")

PPE =  balsheet[balsheet['label'] == 'Property Plant and Equipment']
PPE
```

```python
repairs_PPE = repairs_PPE.merge(PPE, on = ['Unnamed: 0', 'mun_name', 'mun_code', 'Province', 'Latitude', 'Longitude'],
                                how = 'left')
repairs_PPE = repairs_PPE[['Unnamed: 0', 'mun_code', 'mun_name', 'Province', 'Latitude', 'Longitude',
                           'repairs_maintenance', 'total_amount']]
repairs_PPE = repairs_PPE.rename(columns = {'total_amount': 'ppe'})

repairs_PPE.head()
```

```python
#Investment Property
Investment_prop = balsheet[balsheet['label'] == 'Investment Property']
Investment_prop.head()
```

```python
repairs_PPE = repairs_PPE.merge(Investment_prop, on = ['Unnamed: 0', 'mun_name', 'mun_code', 'Province', 'Latitude', 'Longitude'],
                                how = 'left')
repairs_PPE = repairs_PPE[['Unnamed: 0', 'mun_code', 'mun_name', 'Province', 'Latitude', 'Longitude',
                           'repairs_maintenance', 'ppe', 'total_amount']]
repairs_PPE = repairs_PPE.rename(columns = {'total_amount': 'Invest_property'})

repairs_PPE.head()
```

```python
repairs_PPE['repairs_PPE_measure'] = repairs_PPE['repairs_maintenance'] / (repairs_PPE['ppe'] + repairs_PPE['Invest_property'])
repairs_PPE.head()
```
