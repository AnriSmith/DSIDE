```python
def loaddata(path):
    """
    load the data
    """ 
    path = path
    all_files = glob.glob(os.path.join(path, "*.csv"))
    data = []
    df = pd.DataFrame()
    for fil in all_files:
        dataload = pd.read_csv(fil,skipinitialspace=True)
        data.append(dataload)
    return data    
```

```python
def wrangle_youth(data):
    """
    Wrangle municipal data
    """
    
    data = data.set_index('financial_year_end.year')
    data.drop(" ")
    data.index.name = None
    
    
    '    ======== Uncomment section not being used to clean data below =====================    '
    
    # ------- Use this for income expenditure, repair and maintanace, cash flow data use for balance sheet ------
    
    data.drop([ 'amount_type.label', 'item.code', 'financial_period.period','item.composition',
            'item.position_in_return_form','period_length.length','item.return_form_structure'],1, inplace=True)

    data.columns = ['total_amount','account_type','mun_code','mun_name', 'label']
    
    
    #------------------------------------ Use it for debitor -----------------------------------
        
    data.drop(['amount_type.label','item.code', 'financial_period.period','item.composition',
            'item.position_in_return_form','period_length.length','item.return_form_structure','l60_amount.sum',
           'g1_amount.sum', 'l30_amount.sum','l180_amount.sum', 'l150_amount.sum', 'l1_amount.sum',
        'l120_amount.sum', 'l90_amount.sum','customer_group.code','badi_amount.sum','bad_amount.sum']
        ,1, inplace=True)

    data.columns  = ['mun_name','mun_code', 'account_type','label','total_amount']
    

  
     # ------------------ Use this for wasteful expenditure data ----------------------------------------------
    
    data.columns = [ 'mun_name','mun_code','item.code', 'label','total_amount']
    data.drop("item.code",1, inplace=True)
    

    
    return data
```

```python
def Filter(data, text, Column):
    """
    Filters table 
    
    text: what you want to filter with
    Column: Column you want to filter in
    """
    word_filter = text
    Column = Column
    data = data[data[Column].str.contains(word_filter, na = False)]
    
    return data
```
